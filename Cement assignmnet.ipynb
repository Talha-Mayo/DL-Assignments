```python
#Stablizing our data column wise

col1_mean=data['Cement (component 1)(kg in a m^3 mixture)'].mean(axis=0)
col1_std=data['Cement (component 1)(kg in a m^3 mixture)'].std(axis=0)
data['Cement (component 1)(kg in a m^3 mixture)']-=col1_mean
data['Cement (component 1)(kg in a m^3 mixture)']/=col1_std

col2_mean=data['Blast Furnace Slag (component 2)(kg in a m^3 mixture)'].mean(axis=0)
col2_std=data['Blast Furnace Slag (component 2)(kg in a m^3 mixture)'].std(axis=0)
data['Blast Furnace Slag (component 2)(kg in a m^3 mixture)']-=col2_mean
data['Blast Furnace Slag (component 2)(kg in a m^3 mixture)']/=col2_std

col3_mean=data['Fly Ash (component 3)(kg in a m^3 mixture)'].mean(axis=0)
col3_std=data['Fly Ash (component 3)(kg in a m^3 mixture)'].std(axis=0)
data['Fly Ash (component 3)(kg in a m^3 mixture)']-=col3_mean
data['Fly Ash (component 3)(kg in a m^3 mixture)']/=col3_std

col4_mean=data['Water  (component 4)(kg in a m^3 mixture)'].mean(axis=0)
col4_std=data['Water  (component 4)(kg in a m^3 mixture)'].std(axis=0)
data['Water  (component 4)(kg in a m^3 mixture)']-=col4_mean
data['Water  (component 4)(kg in a m^3 mixture)']/=col4_std


col5_mean=data['Superplasticizer (component 5)(kg in a m^3 mixture)'].mean(axis=0)
col5_std=data['Superplasticizer (component 5)(kg in a m^3 mixture)'].std(axis=0)
data['Superplasticizer (component 5)(kg in a m^3 mixture)']-=col5_mean
data['Superplasticizer (component 5)(kg in a m^3 mixture)']/=col5_std

col6_mean=data['Coarse Aggregate  (component 6)(kg in a m^3 mixture)'].mean(axis=0)
col6_std=data['Coarse Aggregate  (component 6)(kg in a m^3 mixture)'].std(axis=0)
data['Coarse Aggregate  (component 6)(kg in a m^3 mixture)']-=col6_mean
data['Coarse Aggregate  (component 6)(kg in a m^3 mixture)']/=col6_std

col7_mean=data['Fine Aggregate (component 7)(kg in a m^3 mixture)'].mean(axis=0)
col7_std=data['Fine Aggregate (component 7)(kg in a m^3 mixture)'].std(axis=0)
data['Fine Aggregate (component 7)(kg in a m^3 mixture)']-=col7_mean
data['Fine Aggregate (component 7)(kg in a m^3 mixture)']/=col7_std

col8_mean=data['Age (day)'].mean(axis=0)
col8_std=data['Age (day)'].std(axis=0)
data['Age (day)']-=col8_mean
data['Age (day)']/=col8_std

```

```python
#Spliting data into data and labels

input_data=data.iloc[:,0:8]
output_labels=data.iloc[:,8:9]
```

```python
input_data
output_labels
```

```python
#Spliting data and labels into test data,test labels, train data, train labels,validation_data and validation_labels.
#5o% for Testing, 20% for validation, 30% for training


training_data=input_data[0:515]
training_labels=output_labels[0:515]

validation_data=input_data[515:721]
validatio_labels=output_labels[515:721]

testing_data=input_data[721:]
testing_labels=output_labels[721:]
```

```python
#Creating model

from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras import Sequential

model=models.Sequential()

model.add(layers.Dense(10, activation='relu',input_shape=(8,)))
model.add(layers.Dense(8,activation='relu'))
model.add(layers.Dense(8,activation='relu'))
model.add(layers.Dense(1,activation='relu'))
```

```python
#Compiling model

model.compile(optimizer='rmsprop',loss='mse', metrics=['mae'])
```

```python
#Fitting data into model

model.fit(training_data,training_labels,epochs=20,validation_data=(validation_data,validatio_labels))

```

```python
#Evaluating model

model.evaluate(testing_data,testing_labels)
```

```python
#Make predictions from test data

model.predict(testing_data)
```
